```python
import cv2
cap = cv2.VideoCapture(r"C:\Users\nisha\OneDrive\Desktop\ML Projects\Kalkini_Project\input_videos\test.mp4")
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Not Frames")
        break
    cv2.imshow("Test Video", frame)
    print("Frames")
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
